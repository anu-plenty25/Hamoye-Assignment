```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
dff = df.columns.to_list()
dff
```

```python
for x in dff:
    df[x] = le.fit_transform(df[x])
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

normalised_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
features_df = normalised_df.drop(columns=['date', 'lights','Appliances'])
energy_target = normalised_df['Appliances']
```

```python
#Now, we split our dataset into the training and testing dataset. Recall that we had earlier segmented the features and target variables.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, energy_target, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

#fit the model to the training dataset
lr.fit(x_train, y_train)

#obtain predictions
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, pred)
round(mae, 2)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, pred)
round(r2_score, 2)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, pred)
round(mse, 2)
```

```python
# to calculate the root mean squared error

rmse = np.sqrt(mse)
round(rmse, 3)
```

```python
#to calculate the residual sum of squares

rss = np.sum(np.square(y_test - pred))
round(rss, 2)
```
